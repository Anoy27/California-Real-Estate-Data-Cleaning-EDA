```python
def standardize_levels(level):
    level = str(level).lower()
    if any(x in level for x in ["0","one", "1", "one story"]):
        return "One Story"
    elif any(x in level for x in ["two", "2", "two story"]):
        return "Two Story"
    elif any(x in level for x in ["three", "tri", "3", "three or more"]):
        return "Three+ Stories"
    elif "split" in level or "multi" in level:
        return "Split/Multi-Level"
    elif "four" in level or "five" in level or "+" in level:
        return "Four+ Stories"
    else:
        return "Other"

# Apply the mapping function
rs['Level'] = rs['levels'].apply(standardize_levels)
```

```python
def standardize_pool(pool):
    pool = str(pool)
    if any(x in pool for x in ["0"]):
        return "Not Available"
    elif any(x in pool for x in ["1"]):
        return "Available"
    else:
        return "Unknown"

# Apply the function
rs['pool'] = rs['pool'].apply(standardize_pool)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=rs, x='livingArea', kde=True, color='purple', bins=20)
plt.title('Histogram with KDE for Outlier Detection', fontsize=14)
plt.show()
```

```python
def find_outliers(group):
    Q1 = group['livingArea'].quantile(0.25)
    Q3 = group['livingArea'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = group[(group['livingArea'] > upper_bound)]
    return outliers
outliers = (
    rs.groupby('Property_Type')[['livingArea']]
    .apply(find_outliers)
)
outliers.head(100)
```
